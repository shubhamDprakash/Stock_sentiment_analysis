```python
headlines = []
for row in range(0,len(df.index)):
    headlines.append(' '.join(str(x) for x in df.iloc[row,1:25]))
df['news'] = headlines
df.drop(news_columns_list, axis='columns', inplace=True)
df.head()
```

```python
# Get the Independent Features
X=df.drop('Label',axis=1)
X.head()
```

# stemming and removing stopwords

```python
messages=X.copy()
```

```python
messages['news'][1]
```

```python
messages.reset_index(inplace=True)
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
# Dataset Preprocessing
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []
for i in range(0, len(messages)):
    if (i%100 == 0):
        print(i)
    review = re.sub('[^a-zA-Z]', ' ', messages['news'][i])
    review = review.lower()
    review = review.split()
    
    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus[0:5]
```

```python
import numpy as np
X_final=np.array(corpus)
y_final=np.array(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size=0.2, shuffle=False)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
```

```python
# implement BAG OF WORDS
countvector=CountVectorizer(ngram_range=(2,2))
traindataset=countvector.fit_transform(X_train)
```

```python
# implement RandomForest Classifier
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(traindataset,y_train)
```

```python
# Predict for the Test Dataset

test_dataset = countvector.transform(X_test)
predictions = randomclassifier.predict(test_dataset)
```

```python
# Import library to check accuracy
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
```

```python
matrix=confusion_matrix(y_test,predictions)
print(matrix)
score=accuracy_score(y_test,predictions)
print(score)
report=classification_report(y_test,predictions)
print(report)
```
